duplikasi values

````python
from fast_ml.feature_selection import get_duplicate_features
dftest = df_kolom.copy()

# mendapatkan kolom-kolom dengan duplikasi values
duplicate_features = get_duplicate_features(dftest)
print('Duplikasi kolom:\n')
print(duplicate_features)

#semua kolom yang duplikasi value sebagai list
duplicate_features_list= duplicate_features.loc[duplicate_features['Desc']=='Duplicate Values','feature2'].to_list()
print('\nList kolom duplikasi values:\n')
print(duplicate_features_list)


#informasi kolom yang sebelum dihapus
print('\nBanyaknya kolom sebelum penghapusan kolom duplikasi values:' + str(dftest.shape[1]))
dftest.drop(columns=duplicate_features_list, inplace=True)
#kolom setelah dihapus
print('Jumlah kolom setelah penghapusan kolom duplikasi value:' + str(dftest.shape[1]))
dftest
````

#**HURUF KAPITAL YANG TIDAK KONsisTEN**

````python
apalagi= pd.DataFrame({'NAMA':['Asep','ASEP','asep']})
print(apalagi)

#ubah menjadi kapital semua
apalagi['NAMA']=apalagi['NAMA'].str.upper()
print('\n')#spasi
print(apalagi)
````

extra whitespace pada data teks

````python
var1= pd.DataFrame({'NAMA':['  asep    ','asep   ','asep rawrr       ']})
print(var1)

#jarakspasi diperbaiki
print('\n')
var1['NAMA']=var1['NAMA'].str.strip()
print(var1)
````

detail teks tidak perlu

````python
asepjadivar= pd.DataFrame({'NAMA':['ASEP (BINTI TATANG)','UDIN (BINTI TATANG)']})
print(asepjadivar)

#ubah detail ga perlu
asepjadivar['NAMA']=asepjadivar['NAMA'].str.replace(r' \(BINTI.*','',regex=True)
print('\n')
print(asepjadivar)
````

penulisan tidak konsisten

````python
a= pd.DataFrame({'alamat':['jln. cihanjuang','jl. barudicor']})
print(a)

#ubah penulisan tidak konsisten
a['alamat']=a['alamat'].str.replace(r'jl.* ', 'jalan ', regex=True)
print('\n')
print(a)
````
