#mengabil beberapa baris tertentu dengan loc 

```python
asep.iloc[2,1] #mengambil baris ke 3 kolom ke 5
```

```python
asep.iloc[[2,22,10],[1,4]] #mengambil baris ke 2 22 10 dan menampilkan data kolom ke 1 dan 4 
```

index : dibelakang bermaskud mengambbil semua data di belakang atau kita bisa namain infinity  

```python
asep.iloc[:10 ,1:]  #data iloc dikosongkan di samping kosong berarti kita mengambbil semua data sampai index ke 10 dan semua data baris dari 1 hiingga selesai  
```

```python
asep.iloc[-2,-2] #ini akan mengambil data dari belakang dimana dibelakang ke dua adalah kota asal yusuf 
```

```python
asep.iloc[-20,-1] #ini akan mengambil data dari 20 tapi dari belakang dan data kolom terakhir yaitu provinsi asal ode
```
